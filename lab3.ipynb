```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import scale, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

%matplotlib inline
```

```python
census = pd.read_csv('swiss-census.csv')
census.head()
```

# Fertility, Agriculture and Catholic

```python
X = census[['Agriculture', 'Catholic', 'Education']]
Y = census['Fertility']

X_train = X[:24]
Y_train = Y[:24]

X_test = X[24:]
Y_test = Y[24:]

linmod = LinearRegression()
linmod.fit(X_train,Y_train)

preds = linmod.predict(X_test)

plt.scatter(Y_test, preds, color='g')
plt.xlabel("Actual Y Value")
plt.ylabel("Predicted Y Value")
plt.show()

mse = mean_squared_error(Y_test, preds)
print('Mean Squared error: ', mse)

rs = r2_score(Y_test, preds)
print('R2 Score:: ', rs)
```

# Agriculture and Catholic

```python
X = census[['Agriculture', 'Catholic']]
Y = census['Fertility']

X_train = X[:24]
Y_train = Y[:24]

X_test = X[24:]
Y_test = Y[24:]

linmod = LinearRegression()
linmod.fit(X_train,Y_train)

preds = linmod.predict(X_test)

plt.scatter(Y_test, preds, color='g')
plt.xlabel("Actual Y Value")
plt.ylabel("Predicted Y Value")
plt.show()

mse = mean_squared_error(Y_test, preds)
print('Mean Squared error: ', mse)

rs = r2_score(Y_test, preds)
print('R2 Score:: ', rs)
```

# Interaction between Agricluture and Catholic

```python
poly = PolynomialFeatures(2, interaction_only=True)

polyX = poly.fit_transform(X)

X_train = polyX[:24]
Y_train = Y[:24]

X_test = polyX[24:]
Y_test = Y[24:]

linmod2 = LinearRegression()
linmod2.fit(X_train,Y_train)

preds = linmod2.predict(X_test)

plt.scatter(Y_test, preds, color='g')
plt.xlabel("Actual Y Value")
plt.ylabel("Predicted Y Value")
plt.show()

mse = mean_squared_error(Y_test, preds)
print('Mean Squared error: ', mse)

rs = r2_score(Y_test, preds)
print('R2 Score:: ', rs)
```
